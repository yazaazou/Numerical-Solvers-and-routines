## Question 1

```python
def monte_for(N):
    n= 0
    for i in range(N):
        x = 2*np.random.rand(1) -1
        y = 2*np.random.rand(1) -1
        a =(x**2 +y**2)
        if  a<=1:
            n=n+1
    return (4*n/N)

def monte_vect(N):
    
    x = 2*np.random.rand(N) -1
    y = 2*np.random.rand(N) -1
    
    return sum(x**2 +y**2 <=1) *4/N
```

```python
N=int(1e6)

start= time.time()
monte_for(N)
end = time.time()
time_for= end-start

start= time.time()
monte_vect(N)
end = time.time()
time_vect= end-start

print("For loop function takes ", time_for, "seconds")
print("The vectorized version takes ",time_vect, "seconds")
```

#### Clearly, there is a noticeable advantage to using the vectorized version of the function; it took the for loop function nearly 4 times as long to perform the same calculation compared to its vectorized counterpart.



## Question 2

```python
def fib():
    f1 = 0
    f2 = 1
    while True:
        yield f1
        f1, f2 = f2, f1+f2

for index, number in enumerate(fib()):
    if len(str(number)) == 1000:
        print("The first Fibonacci number to contain 1000 digits has index n =",index)
        break
```

## Question 3


```python
relative=[]
powers=[]
for i in range(1,14) :
    p = 10**i
    powers.append(p)
    
    num_exp= (1+(1/p))**p 
    relative.append( abs(np.exp(1) - num_exp)/(sp.exp(1)) )
    

    
```

```python
fig, axes = plt.subplots(figsize=(14,8))
plt.plot(powers,relative)
plt.xscale("log")
plt.xlabel("Powers")
plt.ylabel("Relative Error")
plt.show()
```

#### As $n$ approaches $\infty$ the calculation approaches the actual value of $e$ since the round-off error is decreasing. However, for $n$ =14 the error stars to increase again since the truncation error is starting to dominate the round-off error.
